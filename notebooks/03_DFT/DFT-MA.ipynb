# DFT: Frequenzgang Moving Average (MA) Filters
In diesem Notebook zur gleichnamigen Übungsaufgabe wird gezeigt, wie man den Frequenzgang eines Moving Average Filters aus dessen Impulsantwort berechnet und darstellt.

```python
from __future__ import print_function, division
import os, sys
module_path = os.path.abspath(os.path.join('..')) # append directory one level up to import path
if module_path not in sys.path:  # ... if it hasn't been appended already
    sys.path.append(module_path)
    
import matplotlib.pyplot as plt
%matplotlib inline
import dsp_fpga_lib as dsp # import and set plot properties
import numpy as np

np.set_printoptions(threshold=20) # for > 20 items, print with (1,2, ... , 99, 100)
import scipy.signal as sig

figsize = {"figsize": (12,5), "facecolor":"w"}
```

Sie wollen aus der Impulsantwort $h_{MA,3} = \{1;1;1;1\}$ eines MA-Filters der Ordnung $N = 3$ dessen Frequenzgang mit Hilfe der DFT bestimmen. Welches Ergebnis erwarten Sie? Sie versuchen dies zunächst ganz naiv mit folgendem Python-Skript:



```python
h = np.ones(4) # Geben Sie hier die Länge des MA-Filters (Ordnung + 1) an.
n = np.arange(len(h))
fig = plt.figure(1, **figsize)
fig.clf()
ax1 = fig.add_subplot(211)
ax1.stem(n,h)
ax1.set_xlabel(r"$n \; \rightarrow$")
ax1.set_ylabel(r"$h[n] \; \rightarrow$")

ax2 = fig.add_subplot(212)
H = np.fft.fft(h)
k = np.arange(len(H))
ax2.stem(k,H)
ax2.set_xlabel(r"$k \; \rightarrow$")
ax2.set_ylabel(r"$H[k] \; \rightarrow$")

fig.tight_layout()
plt.show()
```
