```python
h = np.ones(8)#[1,1,1,1]
n = np.arange(len(h))
plt.figure(1);
plt.subplot(211)
plt.stem(n,h)
plt.ylabel(r"$h[n] \; \rightarrow$")
plt.xlabel(r"$n \; \rightarrow$")

plt.subplot(212)
H = np.fft.fft(h,256)
k = np.arange(len(H))
plt.stem(k,H)

plt.xlabel(r"$k \; \rightarrow$")
plt.ylabel(r"$|H[k]| \; \rightarrow$")
plt.tight_layout()
plt.show()
```
